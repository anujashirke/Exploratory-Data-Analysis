```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file_path="C:/Users/91897/Desktop/Desktop/Anuja/The Sparks Foundation/SampleSuperstore.csv"
mydata=pd.read_csv(file_path)
```

```python
mydata1=mydata.drop_duplicates(inplace=False)
```

```python
mydata1.isnull().count()
```

```python
mydata1.describe()
```

```python
plt.figure(figsize=(7,10))
plt.pie(mydata1['Category'].value_counts(), labels=mydata1['Category'].value_counts().index, autopct="%1.1f%%");
```

Above pie chart shows, the super store stocks more of office supplies items, moderate amount of furniture and few of technology items.

```python
sns.jointplot(x='Discount', y='Profit', data=mydata1, kind='scatter');
```

The jointplot interprets that discount is inversely proportional to profit. We can see as the discount rate increases profit goes down. When the retailer sell's an item without discount then the profit rate is highest.

```python
ct1=pd.crosstab(mydata1['Sub-Category'], mydata1['Category'])
ct1
```

```python

ct1.plot.bar(stacked=True);
```

The above plot show the count of sub categories among categories. We can see binders, paper, storage and art are in huge amount i.e. consumers/corporates of the USA purchase such office supplies in greater amount. Among technology, customers demand for Accessories and Phones. Customers prefer Furnishings and Chairs among furniture.


```python
ct2=pd.crosstab(mydata1['State'], mydata1['Category'])
ct2
```

```python
fig,ax=plt.subplots(figsize=(12,10))
ct2.plot.bar(stacked=True,ax=ax);
```

The above plot talks about sell of each item in different States. Office Supplies are more often sold in California, New York and Texas. Here we can interpret that MNN companies and offices are located in these states and hence there's more amount of office supplies.
Furniture is sold on a moderate level in California, New York, Texas and Washington.
Technology is supplied to California, New York, Texas, Ohio and Pennsylvania.
Here overall analysis can say that retailer supply more items to California, New York and Texas i.e. the demand of such products is high. 

```python
plt.scatter(x=mydata1['Sales'], y=mydata1['Profit']);
plt.xlabel('Sales');
plt.ylabel('Profit');
```

Here the scatter plot talks about the distribution of Sales and Profit. We can see Profit and Sales are directly proportional to each other. There are some exceptional values that show less profit but more sales. 
We can conclude that the retail store owner usually earn profit, the exceptions could be off-season or due to some global crisis. 

```python
fig, ax=plt.subplots(figsize=(12,10))
sns.stripplot(mydata1['Profit'], mydata1['State'], ax=ax, jitter=True);
```

The strip plot shows distribution of Profit across State. Profit rates are high in New York compare to others. The is denser at California and New York i.e., more profit is earned by selling products in these two states. 
Some states do show loss in the business (Texas, North California, Florida, Nebraska, Pennsylvania, Colorado, Ohio, Illinois). Here the shop owner can organize a survey to know what people really want or can conduct a business research.
